```python
with open("api_key.txt", "r") as file:
    api_key = file.read()
```

```python
for flower in ['daisy', 'dandelion', 'rose', 'sunflower', 'tulip']:
    for i in range(1, 11):
        print(i)
        import requests
        import os

        # Pexels endpoint for searching photos
        SEARCH_URL = 'https://api.pexels.com/v1/search'

        # Headers for authentication
        headers = {
            'Authorization': api_key
        }

        # Parameters for the search query
        params = {
            'query': f'{flower}',  # Change this to whatever you're searching for
            'per_page': 80,  # Number of results per page
            'page': i  # Page number
        }

        folder_name = f'{flower}'
        if not os.path.exists(folder_name):
            os.makedirs(folder_name)

        # Make the GET request to Pexels API
        response = requests.get(SEARCH_URL, headers=headers, params=params)

        # Check if the request was successful
        if response.status_code == 200:
            # Convert the response to JSON
            data = response.json()

            # Iterate through each photo in the response
            for photo in data['photos']:
                # Get the URL of the photo
                photo_url = photo['src']['original']
                
                # Get the photo ID (to use as filename)
                photo_id = photo['id']
                
                # Download the photo
                photo_response = requests.get(photo_url)
                
                # Check if the download was successful
                if photo_response.status_code == 200:
                    # Write the photo to a file in the specified folder
                    file_path = os.path.join(folder_name, f'{photo_id}.jpg')
                    with open(file_path, 'wb') as file:
                        file.write(photo_response.content)
                    print(f'Downloaded photo {photo_id} to {file_path}')
                else:
                    print(f'Failed to download photo {photo_id}')
        else:
            print('Request failed: ', response.status_code)
```
